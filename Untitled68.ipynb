## Load and Merge Datasets


```python
import pandas as pd

# Load the datasets
order_items_df = pd.read_csv('/content/olist_order_items_dataset[1].csv')
payments_df = pd.read_csv('/content/olist_order_payments_dataset.csv')

# Merge the datasets on 'order_id'
merged_df = pd.merge(order_items_df, payments_df, on='order_id', how='inner')

# Display the first few rows of the merged DataFrame
print("Merged DataFrame head:")
print(merged_df.head())
```

## Calculate Total Revenue


```python
total_revenue = merged_df['payment_value'].sum()
print(f"Total Revenue: {total_revenue:.2f}")
```

## Calculate Average Order Value (AOV)


```python
unique_orders = merged_df['order_id'].nunique()
aov = total_revenue / unique_orders
print(f"Average Order Value (AOV): {aov:.2f}")
```

## Analyze Revenue Over Time


```python
merged_df['shipping_limit_date'] = pd.to_datetime(merged_df['shipping_limit_date'])
merged_df_time = merged_df.set_index('shipping_limit_date')
monthly_revenue = merged_df_time['payment_value'].resample('M').sum()

print("Monthly Revenue:")
print(monthly_revenue.head())
```

```python
merged_df['shipping_limit_date'] = pd.to_datetime(merged_df['shipping_limit_date'])
merged_df_time = merged_df.set_index('shipping_limit_date')
monthly_revenue = merged_df_time['payment_value'].resample('ME').sum()

print("Monthly Revenue:")
print(monthly_revenue.head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(monthly_revenue.index, monthly_revenue.values, marker='o', linestyle='-')
plt.title('Monthly Revenue Over Time')
plt.xlabel('Date')
plt.ylabel('Revenue (BRL)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
