## Calculate Total Revenue and AOV

### Subtask:
Calculate the Total Revenue by summing all payment values and the Average Order Value (AOV) by averaging payment values per order from the loaded data.


**Reasoning**:
To calculate the Total Revenue and Average Order Value (AOV), I will sum the 'payment_value' column for Total Revenue, count the unique 'order_id' values for the number of orders, and then divide the Total Revenue by the number of unique orders to get the AOV.



```python
total_revenue = df_payments['payment_value'].sum()
num_unique_orders = df_payments['order_id'].nunique()
aov = total_revenue / num_unique_orders

print(f"Total Revenue: {total_revenue:,.2f}")
print(f"Number of Unique Orders: {num_unique_orders}")
print(f"Average Order Value (AOV): {aov:,.2f}")
```
